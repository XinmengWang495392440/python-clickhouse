```python
import pandas as pd, numpy as np
import random
import time
```

```python
a1=(2018,1,1,0,0,0,0,0,0)            
a2=(2018,12,31,23,59,59,0,0,0)   

cust_id = np.random.randint(2000000000,5000000000,500000)
fraud_label = np.random.randint(0,2,500000)
risk_level = np.random.randint(0,4,500000)

## get random time
start=time.mktime(a1)    
end=time.mktime(a2)   

date_list = []
for i in range(500000):      
    t=random.randint(start,end)    
    date_touple=time.localtime(t)        
    date=time.strftime("%Y-%m-%d",date_touple)  
    date_list.append(date)

## get random gender
gender_list = []
gender = ['Male', 'Female', '~']

for i in range(500000):
    gender_random = ''.join(random.sample(gender,1))
    gender_list.append(gender_random)
    
## get random jobs
job_list = []
job = ['Students', 'Accounter', '~', 'Engineer', 'Cook', 'Teacher', 'Artist', 'Professional Manager']

for i in range(500000):
    job_random = ''.join(random.sample(job,1))
    job_list.append(job_random)
```

```python
basic_table = pd.DataFrame({
        'cust_id': pd.Series(
            (cust_id),
            dtype='object'
        ),
        'date':pd.Series(
            date_list
        )
    })
```

```python
fraud_case_table = pd.DataFrame({
        'cust_id': pd.Series(
            (cust_id),
            dtype='object'
        ),
        'date':pd.Series(
            date_list
        ),
        'fraud_label':pd.Series(
            (fraud_label),
            dtype='uint8'
        ),
        'risk_level':pd.Series(
            (risk_level),
            dtype='uint8'
        ),
    })
```

```python
account_table = pd.DataFrame({
        'cust_id': pd.Series(
            (cust_id.repeat(10)),
            dtype='object'
        ),
        'date':pd.Series(
            np.asarray(date_list).repeat(10)
        ),
        'amount':pd.Series(
            np.random.randint(0,1000000,5000000)
        ),    
    })
```

```python
information_table = pd.DataFrame({
        'cust_id': pd.Series(
            (cust_id),
            dtype='object'
        ),
        'date':pd.Series(
            (date_list),
            dtype = 'object'
        ),
        'gender':pd.Series(
            (gender_list),
            dtype='object'
        ),
        'job':pd.Series(
            (job_list),
            dtype='object'
        ),
    })
```

```python
basic_table.to_csv('fake_data/basic_table.csv', index = None, header = False)
fraud_case_table.to_csv('fake_data/fraud_case_table.csv', index = None, header = False)
account_table.to_csv('fake_data/account_table.csv', index = None, header = False)
information_table.to_csv('fake_data/information_table.csv', index = None, header = False)
```
